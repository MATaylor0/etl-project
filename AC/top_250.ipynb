```python
# Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from unicodedata import normalize
import requests as req
from config import api_key
from pprint import pprint
import requests
import json
from datetime import datetime
import numpy as np
```

```python
# Create list of top 250 IMBD movies to loop through 
table_MN = pd.read_html('https://www.imdb.com/chart/top/?ref_=nv_mv_250')
top_250=table_MN[0]["Rank & Title"].str[4:-6]
top_250=list(top_250)
```

```python
# Create base
base_url = "http://www.omdbapi.com/?t="
api_key = "&apikey=" + api_key

# Set up lists to hold response data
Title=[]
Year=[]
Rated=[]
Released=[]
Runtime=[]
Genre=[]
Director=[]
Writer=[]
Actors=[]
Plot=[]
Language=[]
Country=[]
Poster=[]
Metascore=[]
imdbID=[]
imdbRating=[]
imdbVotes=[]
Type=[]
Production=[]
```

```python
# Loop through movies and add info to lists 
for movie in top_250:
    target_url = base_url + movie + api_key
    response = requests.get(target_url).json()
    
    Title.append(response.get('Title'))
    Year.append(response.get('Year'))
    Rated.append(response.get('Rated'))
    Released.append(response.get('Released'))
    Runtime.append(response.get('Runtime'))
    Genre.append(response.get('Genre'))
    Director.append(response.get('Director'))
    Writer.append(response.get('Writer'))
    Actors.append(response.get('Actors'))
    Plot.append(response.get('Plot'))
    Language.append(response.get('Language'))
    Country.append(response.get('Country'))
    Poster.append(response.get('Poster'))
    Metascore.append(response.get('Metascore'))
    imdbID.append(response.get('imdbID'))
    imdbRating.append(response.get('imdbRating'))
    imdbVotes.append(response.get('imdbVotes'))
    Type.append(response.get('Type'))
    Production.append(response.get('Production'))
```

```python
# Create dataframe to hold data
top_250_df = pd.DataFrame({
    "Title":Title,
    "Year":Year,
    "Rated":Rated,
    "Released":Released,
    "Runtime":Runtime,
    "Director":Director,
    "Plot":Plot,
    "Poster":Poster,
    "Metascore":Metascore,
    "imdbID":imdbID,
    "imdbRating":imdbRating,
    "imdbVotes":imdbVotes,
    "Type":Type
})
```

```python
# Check for null value rows
top_250_df[top_250_df.isnull().any(axis=1)]

top_250_df.replace("N/A", np.NaN, inplace = True)

top_250_df["imdbID"] = top_250_df["imdbID"].drop_duplicates()

# Drop rows
top_250_df.dropna(how='any', inplace = True)
```

```python
# Adding a rank column by resetting the index
top_250_df.reset_index(inplace = True)
top_250_df.head()
```

```python
# Increasing the value of the rank column by 1
top_250_df.rename(columns = {"index": "Rank"}, inplace = True)
top_250_df["Rank"] = top_250_df["Rank"].apply(lambda x: x + 1)
top_250_df.head()
```

```python
# Removing the "min" from the Runtime string
top_250_df["Runtime"] = top_250_df["Runtime"].str[:-3]
top_250_df.head()
```

```python
# Formatting the date to be read by SQL
top_250_df["Released"] = top_250_df["Released"].apply(lambda x: datetime.strptime(x, "%d %b %Y").strftime("%Y-%m-%d"))
top_250_df.head()
```

```python
top_250_df["imdbVotes"] = top_250_df["imdbVotes"].str.replace(",","")
top_250_df.head()
```

```python
# Export to csv
top_250_df.to_csv("Top_250_Movies.csv", index=False, header=True)
```

```python
top_250_genre_df = top_250_genre_df.loc[top_250_genre_df["imdbID"].isin(top_250_df["imdbID"])]
```

```python
# Checking for null values
top_250_genre_df.isnull().sum()
```

```python
# Dropping any null values and resetting the index
top_250_genre_df.replace("N/A", np.NaN, inplace = True)

top_250_genre_df.dropna(how = "any", inplace = True)

top_250_genre_df.reset_index(drop = True, inplace = True)
```

```python
# Creating empty lists to hold information for new dataframe
id_list = []
title_list = []
genre_list = []

# Looping through the dataframe to seperate data into first normal form
for i in range(0, len(top_250_genre_df)):
    if len(top_250_genre_df["Genre"][i].split(",")) > 1:
        words = top_250_genre_df["Genre"][i].split(",")
        result = list(map(lambda x: x.strip(), words))
        
        genre_list.extend(result)
        
        for j in range(0, len(top_250_genre_df["Genre"][i].split(","))):
            id_list.append(top_250_genre_df["imdbID"][i])
            title_list.append(top_250_genre_df["Title"][i])
    else:
        words = top_250_genre_df["Genre"][i].strip()
        
        genre_list.append(words)
        
        id_list.append(top_250_genre_df["imdbID"][i])
        title_list.append(top_250_genre_df["Title"][i])
```

```python
# Creating the new genre dataframe in first normal form
genre_df = pd.DataFrame({
    "imdbID": id_list,
    "Title": title_list,
    "Genre": genre_list
})

genre_df
```

```python
genre_df.reset_index(inplace = True)
```

```python
genre_df.rename(columns = {"index": "id"}, inplace = True)
genre_df
```

```python
# Exporting as a CSV
genre_df.to_csv("top_250_genres.csv", index=False, header=True)
```
